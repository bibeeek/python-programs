```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
y=[]
for x in range(2,200):
    if x%2==0:
        
        y.append(x)
print("displaying even numbers from 2 to 200")
print("\n")
print(y)
```

```python

plt.bar(y,y)
plt.title("Bar of even number from 2 to 200")
plt.show()
```
